## Q0. Import Libraries

```python
# answer here
import matplotlib.pyplot as plt
import pandas as pd
```

## Q1. Create a List
Create a Python list named `numbers` with **five** numeric values of your choice. Then print the list.
- Name the variable exactly `numbers`.
- Use integers only.


```python
# answer here
numbers = [1,2,3,4,5]
print(numbers)
```

## Q2. Line Chart from List
Using the `numbers` list from Q1, plot a **line chart**.
- Use `plt.plot(numbers)`
- Add a title "Simple Line Chart"
- Label x-axis as `Index` and y-axis as `Value`
- Show the plot


```python
# answer here
plt.plot(numbers)
plt.title("Simple Line Chart")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

## Q3. Bar Chart from List
Plot a **bar chart** using the same `numbers` list.
- X-axis should be the index positions of the list (0 to length-1)
- Add a title "Simple Bar Chart"
- Label axes appropriately
- Show the plot


```python
# answer here
plt.bar(range(len(numbers)), numbers)
plt.title("Simple Bar Chart")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

## Q4. Pie Chart from List
Plot a **pie chart** using the `numbers` list.
- Provide labels for each slice (e.g., A, B, C, D, E)
- Show percentage on slices using `autopct="%1.1f%%"`
- Add a title "Simple Pie Chart"
- Show the plot


```python
# answer here
labels = ["A", "B", "C", "D", "E"]
plt.pie(numbers, labels=labels, autopct="%1.1f%%")
plt.title("Simple Pie Chart")
plt.show()

```

## Q5. Create a Pandas Series
Create a Pandas **Series** named `sales` with five values (e.g., monthly sales). Use month names as the index: `Jan`, `Feb`, `Mar`, `Apr`, `May`.
- Print the Series


```python
# answer here
sales = pd.Series([100,200,300,400,500], index=["jan", "feb", "mar", "apr", "may"])
print(sales)
```

## Q6. Line Chart from Series
Plot a **line chart** from the `sales` Series.
- Use `sales.plot(kind="line", title="Monthly Sales")`
- Label x-axis as `Month` and y-axis as `Sales`
- Show the plot


```python
# answer here
sales.plot(kind="line", title="Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Q7. Create a DataFrame
Create a **DataFrame** named `df` with three columns: `Month`, `Sales`, `Profit`.
- `Month`: `Jan`, `Feb`, `Mar`, `Apr`, `May`
- `Sales`: choose five integers
- `Profit`: choose five integers
- Print the DataFrame


```python
# answer here
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [100,200,300,400,500],
    "Profit": [10,20,30,40,50]
})
print(df)
```

## Q8. Line Chart Between Two Columns
Using `df`, plot **two line charts on the same axes**: one for `Sales` and one for `Profit` against `Month`.
- Use `plt.plot(df["Month"], df["Sales"], label="Sales")`
- Use `plt.plot(df["Month"], df["Profit"], label="Profit")`
- Add title "Sales vs Profit"
- Label axes and add legend
- Show the plot


```python
# answer here
plt.plot(df["Month"], df["Sales"], label="Sales")
plt.plot(df["Month"], df["Profit"], label="Profit")
plt.title("Sales vs Profit")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.show()
```

## Q9. GroupBy Setup
Create a new DataFrame named `df_group` with two columns: `Category` and `Sales`.
- Include at least 7 rows
- Use categories like `A`, `B`, `C` (repeated)
- Use integers for `Sales`
- Print the DataFrame


```python
# answer here
df_group = pd.DataFrame({
    "Category": ["A", "B", "C", "A", "B", "C", "A"],
    "Sales": [100,200,300,400,500,600,700]
})
print(df_group)

```

## Q10. GroupBy and Bar Chart
Group `df_group` by `Category` and sum the `Sales`. Store the result in a variable named `grouped`.
- Print `grouped`
- Plot a **bar chart** from `grouped` with title "Total Sales by Category"
- Label axes and show the plot


```python
# answer here
grouped = df_group.groupby("Category").sum()
print(grouped)
grouped.plot(kind="bar", title="Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```
