## Question 1: Comparing Two Lists Using a Line Plot

Problem Statement:
You are given two lists representing monthly sales of two products over 6 months.
You need to visualize how their sales trends compare over time using a line plot.

Approach:
1. Create two lists: months and sales.
2. Use `sns.lineplot()` to plot both lines on the same graph.
3. Add appropriate labels and a title to make the visualization clear.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Example lists
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [45, 55, 60, 70, 65, 75]
sales_B = [40, 50, 58, 68, 70, 80]

# Your line plot code here
sns.lineplot(x=months, y=sales_A, label='Product A')
sns.lineplot(x=months, y=sales_B, label='Product B')

```
